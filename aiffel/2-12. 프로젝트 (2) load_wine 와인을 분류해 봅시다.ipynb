```python
# (1) 모듈 import 하기
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
wine = load_wine()
```

```python
wine.keys()
```

```python
wine_data = wine.data
wine_label = wine.target
```

```python
wine.target_names
```

```python
import pandas as pd

wind_df = pd.DataFrame(data = wine_data, columns=wine.feature_names)
wind_df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(wine_data,
                                                   wine_label,
                                                   test_size=0.2,
                                                   random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier  # DecisionTree
from sklearn.ensemble import RandomForestClassifier #RF
from sklearn import svm                          # svm
from sklearn.linear_model import SGDClassifier   #SGD
from sklearn.linear_model import LogisticRegression  #logistics

dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
svm_model = svm.SVC()
sgd_model = SGDClassifier()
lr_model = LogisticRegression()

# dt 적용
dt_model = dt_model.fit(X_train,y_train)
y_dt_pred = dt_model.predict(X_test)

# rf 적용
rf_model = rf_model.fit(X_train, y_train)
y_rf_pred = rf_model.predict(X_test)

# svm 적용
svm_model = svm_model.fit(X_train, y_train)
y_svm_pred = svm_model.predict(X_test)

# sgd 적용
sgd_model = sgd_model.fit(X_train, y_train)
y_sgd_pred = svm_model.predict(X_test)

# LR 적용
lr_model = lr_model.fit(X_train, y_train)
y_lr_pred = lr_model.predict(X_test)
```

```python
print("DT classification_report ")
print()
print(classification_report(y_dt_pred, y_test))

print("RF classification_report ")
print()
print(classification_report(y_rf_pred, y_test))

print("SVM classification_report ")
print()
print(classification_report(y_svm_pred, y_test))

print("SGD classification_report ")
print()
print(classification_report(y_sgd_pred, y_test))

print("LR classification_report ")
print()
print(classification_report(y_lr_pred, y_test))
```
