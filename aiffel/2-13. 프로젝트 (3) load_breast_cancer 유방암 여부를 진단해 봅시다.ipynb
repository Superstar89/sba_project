```python
# (1) 모듈 import 하기
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
```

```python
breast_cancer = load_breast_cancer()
breast_cancer.keys()
```

```python
breast_cancer.target_names
```

```python
breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target
breast_cancer_df = pd.DataFrame(breast_cancer_data, columns=breast_cancer.feature_names)
breast_cancer_df.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data,
                                                    breast_cancer_label,
                                                   test_size=0.2,
                                                   random_state=2)
```

```python
from sklearn.tree import DecisionTreeClassifier  # DecisionTree
from sklearn.ensemble import RandomForestClassifier #RF
from sklearn import svm                          # svm
from sklearn.linear_model import SGDClassifier   #SGD
from sklearn.linear_model import LogisticRegression  #logistics

dt_model = DecisionTreeClassifier()
rf_model = RandomForestClassifier()
svm_model = svm.SVC()
sgd_model = SGDClassifier()
lr_model = LogisticRegression()


dt_clf = dt_model.fit(X_train, y_train)
y_dt_pred = dt_clf.predict(X_test)

rf_clf = rf_model.fit(X_train, y_train)
y_rf_pred = rf_clf.predict(X_test)

svm_clf = svm_model.fit(X_train, y_train)
y_svm_pred = svm_clf.predict(X_test)

sgd_clf = sgd_model.fit(X_train, y_train)
y_sgd_pred = sgd_clf.predict(X_test)

lr_clf = lr_model.fit(X_train, y_train)
y_lr_pred = lr_clf.predict(X_test)
```

```python
print("DT classification_report ")
print()
print(classification_report(y_dt_pred, y_test))

print("RF classification_report ")
print()
print(classification_report(y_rf_pred, y_test))

print("SVM classification_report ")
print()
print(classification_report(y_svm_pred, y_test))

print("SGD classification_report ")
print()
print(classification_report(y_sgd_pred, y_test))

print("LR classification_report ")
print()
print(classification_report(y_lr_pred, y_test))
```
