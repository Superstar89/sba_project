```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
```

```python
NBA_FILE_PATH = 'datasets/NBA_player_of_the_week.csv'

nba_player_of_the_week_df = pd.read_csv(NBA_FILE_PATH)
```

```python
nba_player_of_the_week_df.head()
```

```python
nba_player_of_the_week_df.describe()
```

```python
heigt_weight_age_df = nba_player_of_the_week_df[['Height CM','Weight KG','Age']]
```

```python
heigt_weight_age_df
```

```python
scaler = preprocessing.MinMaxScaler() # 데이터를 1과 0 사이로 바꿔주는 도구
normalized_data = scaler.fit_transform(heigt_weight_age_df)
normalized_data
```

```python
pd.DataFrame(normalized_data, columns = ['Height', 'Weight', 'Age'])
```

```python
normalized_df = pd.DataFrame(normalized_data, columns = ['Height', 'Weight', 'Age'])
```

```python
normalized_df.describe()
```
