## 20.2 3의 배수일 때와 5의 배수일 때 처리하기

```python
for i in range(1, 101):
    if i % 3 == 0:
        print('Fizz')
    if i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```

## 20.3 3과 5의 공배수 처리하기

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 ==0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(i)
```
